```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('fran_val.csv').rename(columns={'Franchise Value in M': 'Value'})
df['Value']= df['Value'].str.replace(',', '').astype(int)
df.describe()
```

```python
import seaborn as sns

df_norm = df

df_norm.iloc[:, 1] = (df.iloc[:, 1] - df.iloc[:, 1].mean()) / df.iloc[:, 1].std()

print(df_norm.describe())
sns.set_palette('BuPu_r')
sns.distplot(df_norm.Value, bins=12)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 5,
            init = 'k-means++',
            max_iter = 1,
            n_init = 1,
            random_state = 0)

# Define the values to which the KMeans model will be applied, replace with desired columns
# generally will want to normalize dataframe
vals = df['Franchise Value in M'].to_numpy().reshape(-1, 1)
print(vals)
# Fit the model to the values
y_means = km.fit(vals)
print(y_means)
```
